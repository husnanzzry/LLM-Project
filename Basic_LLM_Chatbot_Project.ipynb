```python
import gradio as gr
from google.genai import types

client = genai.Client(api_key="geminiAPIKey")

def chatbot(message, history):

    conversation = ""

    for user_msg, bot_msg in history[-5:]:
        conversation += f"User: {user_msg}\nAssistant: {bot_msg}\n"

    conversation += f"User: {message}"

    try:
        response = client.models.generate_content(
            model="gemini-2.5-flash",
            contents=conversation,
            config=types.GenerateContentConfig(
                temperature=0.8
            )
        )
        return response.text
    except Exception as e:
        return f"Error: {e}"

demo = gr.ChatInterface(                            #A ready-made chatbot UI.
    fn=chatbot,                                     #Most important. Means, when user sends message, call chatbot().
    title="My First LLM Chatbot",
    description="Ask me anything!"
)

demo.launch(share=True)
```
